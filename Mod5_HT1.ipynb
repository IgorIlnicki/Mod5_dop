```python
import pandas as pd
import pandas as pd

url = "2017_jun_final.csv"
table = pd.read_html('https://docs.google.com/spreadsheets/d/1bclX5PbqE8zR1-wOSM3ZTFa53sIC7RXqwty7Abmw9yQ/edit#gid=1724592523', encoding='utf-8')
desired_table = table[0]
# Виведення перших декількох рядків для перевірки
print(desired_table.head())
rows, columns = desired_table.shape
print("\nКількість рядків:", rows)
print("Кількість стовпців:", columns)

print (desired_table.dtypes)
```

```python
# Порахувати кількість пропущених значень в кожному стовпці
missing_counts = desired_table.isnull().sum()
print(missing_counts)
# Розмір таблиці
total_rows = len(desired_table)

# Порахувати частку пропущених значень в кожному стовпці
missing_fraction = missing_counts / total_rows

# Вивести результат
print("Частка пропусків в кожній колонці:")
print(missing_fraction)
```

```python
# Видалення стовпців з пропусками, крім стовпця "Мова програмування"
data_cleaned = desired_table.dropna(subset=[col for col in desired_table.columns if col != 'Мова програмування'])

# Виведення перших декількох рядків для перевірки результату
print(data_cleaned.head())
```

```python
# Перевірка, що залишився тільки стовпець "Мова програмування"
print("Залишився стовпець 'Мова програмування':")
print(data_cleaned.columns)

# Порахувати кількість пропущених значень в кожному стовпці
missing_counts = data_cleaned.isnull().sum()

# Розмір таблиці
total_rows = len(data_cleaned)

# Порахувати частку пропущених значень в кожному стовпці
missing_fraction = missing_counts / total_rows

# Вивести результат
print("Частка пропусків в кожній колонці:")
print(missing_fraction)
```
